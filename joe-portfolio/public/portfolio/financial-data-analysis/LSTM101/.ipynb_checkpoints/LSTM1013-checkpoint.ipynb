```python
import yfinance as yf
import pandas as pd

# 选择黄金资产（你可以改成 "XAUUSD=X" 获取现货黄金）
ticker = "GLD"

# 设置时间范围
start_date = "2013-01-01"
end_date = "2022-12-31"

# 从yfinance获取数据
data = yf.download(ticker, start=start_date, end=end_date)

# 打印前几行，确认数据结构
print(data.head())

# 保存为CSV
output_file = f"{ticker}_gold_prices_2013_2022.csv"
data.to_csv(output_file, encoding="utf-8-sig")

print(f"数据已保存为: {output_file}")
```

```python
NumCols = df4.columns.drop(['Date'])
NumCols
```

```python
df[NumCols] = df[NumCols].replace({',':''}, regex = True)
df[NumCols]
# 疑问点
```

```python
df[NumCols].iloc[2:].reset_index(drop = True)
```

有些数字可能是以字符串形式存的，比如 "1,234.56"（里面有逗号）。\

.replace({',': ''}, regex=True) 就是把逗号 , 去掉，变成 "1234.56"。\

这样后续才能转换成真正的数值(float)，方便做计算。

```python
df11 = df[NumCols].iloc[2:].reset_index(drop = True)
df12 = df11.astype('float64')
df12
```

df[NumCols] = df[NumCols].astype(float)   # 先转成数字\
df[NumCols] = df[NumCols].round(2)        # 保留两位小数


```python
df_gold_price_cleaned = pd.concat([df4[["Date"]], df12], axis=1)
df_gold_price_cleaned
```
