
# Types of gates in LSTM


## input gates
definition: decide which information to store in the memory cell.It is trained to open when the input is important and close when it is not

## forget gates
definition: decide which information to discard from the memory cell. It is trained to open when the infor is no longer important and close when it is.

## output gate
definition: decide which information to use for the output of LSTM. It is trained to open when the infor is important and close when it is not
