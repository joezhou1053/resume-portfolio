```python
import os
os.environ['KERAS_BACKEND'] = 'tensorflow'  # 必须在导入keras前设置
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_percentage_error
import tensorflow as tf
from tensorflow import keras
from keras import Model
from keras.layers import Input, Dense, Dropout, LSTM
```

```python
df = pd.read_csv("Gold_Price_Cleaned.csv", index_col = 0)
```

```python
df
```

```python
df["Close_growth_rate"] = df["Close"].pct_change() # (t0-t-1)/t-1
df = df.dropna(subset=["Close_growth_rate"])
df
```

```python
df['order'] = [1 if sig>0 else -1 for sig in df["Close_growth_rate"]]
df
```

```python
df['profit'] = df['Close'] * df['order']
df
```

```python
df['wealth'] = df['profit'].cumsum()
df
```

```python
print("total profit made:", df['profit'].sum())
```
